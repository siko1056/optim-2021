# UM03

Consider the well-known [Rosenbrock function](https://en.wikipedia.org/wiki/Rosenbrock_function)

$$
f(x,y) = (a - x)^2 + b \left( y - x^2 \right)^2.
$$

It has a global minimum at $P_{1} = (a, a^2)$ with $f(P_{1}) = 0$.
In this exercise the parameters $a = 1$ and $b = 100$ are chosen,
as in many books on mathematical optimization.

$$
\nabla f = \begin{pmatrix} -2(a - x) - 4bx(y - x^2) \\ 2b(y - x^2) \end{pmatrix}
$$

$$
\nabla^{2} f = \begin{pmatrix} 2 - 4by + 12bx^2 & -4bx \\ -4bx & 2b \end{pmatrix}
$$

**1. Create a plot of $f$ for $-2 \leq x \leq 2$ and $-1 \leq y \leq 4$.**

```octave
f = @(x,y) 100 * (y - x.^2).^2 + (1 - x).^2;
[x, y] = meshgrid (linspace (-2, 2, 30), linspace (-1, 4, 30));
mesh (x, y, f(x,y));
grid on;
xlabel ('x');
ylabel ('y');
zlabel ('f(x,y)');
title ('Rosenbrock function');
hold on;
plot3 (1, 1, 0, 'ro');
view (-30, 50);
```
